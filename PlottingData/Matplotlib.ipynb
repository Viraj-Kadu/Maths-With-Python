```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# 1. Create some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x) 
```

```python
# 2. Plot a line chart
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='b', linestyle='--')
plt.plot(x, z, label='cos(x)', color='r')
plt.title('Sine and Cosine Functions')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Create a bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

plt.figure(figsize=(7, 5))
plt.bar(categories, values, color='g')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# 4. Create a scatter plot
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

plt.figure(figsize=(7, 5))
plt.scatter(x_scatter, y_scatter, c='blue', alpha=0.5)
plt.title('Scatter Plot Example')
plt.xlabel('X Random Values')
plt.ylabel('Y Random Values')
plt.grid(True)
plt.show()
```
